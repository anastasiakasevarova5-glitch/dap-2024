1. Что такое медиана? Рассчитать медиану для предложенного ряда чисел.

Медиана — это число, которое находится в середине упорядоченного набора данных. Она делит выборку на две равные части: половина значений меньше медианы, половина — больше.

- Если количество элементов нечетное, медиана — это элемент в центре.
- Если количество элементов четное, медиана — это среднее арифметическое двух центральных элементов.

Расчет для ряда [1, 4, 2, 5, 3]:

1) Упорядочиваем ряд: [1, 2, 3, 4, 5].
2) Считаем элементы: n=5 (нечетное).
3) Находим центральный элемент: Индекс (5+1)/2=3. Третий элемент — это 3.

2. Что такое дисперсия? Рассчитать дисперсию для предложенного ряда чисел.

Дисперсия (обозначается σ^2 или D) — это мера разброса значений случайной величины относительно её среднего значения. Она показывает, насколько "далеко" в среднем числа в наборе данных отстоят от среднего арифметического.

Расчет для ряда [1, 4, 2, 5, 3]:
1. Находим среднее (1+4+2+5+3)/5=15/5=3.

2. Находим квадраты отклонений от среднего:

(1−3)^2=(−2)^2=4
(4-3)^2=(1)^2=1
(2-3)^2=(-1)^2=1
(5-3)^2=4
(3-3)^2=0

3. Суммируем квадраты отклонений:
4+1+1+4+0=10.

4. Делим на количество элементов (n): 10/5=2.

Примечание: В статистике для генеральной совокупности делят на n, а для выборки — на n−1. Здесь расчет для всей совокупности.

3. Что такое среднеквадратичное отклонение? Рассчитать среднеквадратичное отклонение для предложенного ряда чисел.

Среднеквадратичное отклонение (или стандартное отклонение, обозначается σ) — это квадратный корень из дисперсии. В отличие от дисперсии, которая измеряется в квадратных единицах, среднеквадратичное отклонение измеряется в тех же единицах, что и исходные данные, что делает его более удобным для интерпретации.

Расчет для ряда [1, 4, 2, 5, 3]:Мы уже нашли дисперсию:
σ^2=2. Следовательно, среднеквадратичное отклонение:

σ≈1.41

4. Что такое Series и чем он отличается от DataFrame

Что такое Series и чем он отличается от DataFrame?

- Series (Серия) — это одномерный помеченный массив, способный хранить данные любого типа (int, float, str и т.д.). Похож на столбец в таблице или на словарь в Python. Имеет индексы и значения.

- DataFrame (Датафрейм) — это двумерная структура данных (таблица), состоящая из набора Series (столбцов), которые все имеют общий индекс. Это основная структура для работы с данными в Pandas.


6. Можно ли выполнить операцию сложения над двумя объектами Series, если они имеют:

- одинаковые типы данных, одинаковые размеры и одинаковые индексы?

Да. Поэлементное сложение.	Идеальный случай. s1 + s2 даст новую Series.

- одинаковые типы данных, одинаковые размеры, но разные индексы?

Да.	Pandas выровняет индексы. Элементы с несовпадающими индексами дадут NaN.

- одинаковые типы данных, разные размеры и одинаковые индексы?

Нет. Ошибка ValueError.	Размеры должны совпадать для поэлементных операций.

- разные типы данных, одинаковые размеры и одинаковые индексы?

Да. С выполнением приведения типов (Upcasting).	Числа + Строки приведут к типу object (строки). Числа + Булевы значения сложатся как числа (True=1, False=0).

- разные типы данных, одинаковые размеры и разные индексы?

Да. Но с NaN и приведением типов.	Сначала произойдет выравнивание индексов (появятся NaN), затем сложение с приведением типов.

- разные типы данных, разные размеры и разные индексы?

Нет. Ошибка ValueError.	Несовпадение размеров блокирует операцию.

9. Чем отличается метод iloc от метода loс

- .loc[] — Доступ по меткам (label-based).

- - Использует имена строк (индекс) и имена столбцов.
- - Включает правую границу (в срезах).
- - Пример: df.loc[0:2, 'Name':'Fare'].

- .iloc[] — Доступ по целочисленному положению (integer-location based).
- - Использует номера строк (0, 1, 2...) и номера столбцов.
- - Работает как обычный срез Python (исключает правую границу).
- - Пример: df.iloc[0:3, :4].
