# Лабораторная работа №6. Линейная регрессия

### 1. Импорт библиотек и загрузка данных

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Загрузка датасета
diabetes = datasets.load_diabetes()
```

### 2. Описание набора данных и наименование признаков

```python
print(diabetes.DESCR)
print("Названия признаков:", diabetes.feature_names)
```

Набор данных содержит 442 записи. Признаки (feature_names) — это медицинские показатели, такие как возраст, пол, индекс массы тела (bmi), среднее артериальное давление (bp) и концентрации различных веществ в крови (s1-s6). Целевая переменная (target) — это количественный показатель прогрессирования заболевания.

### 3. Подготовка данных

Создадим DataFrame и выведем первые 5 строк.

```python
# Создаем DataFrame из данных и целевой переменной
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target

# Выводим первые 5 строк
print(df.head())
```

Вы увидите таблицу с 10 признаками и столбцом `target`, содержащую числовые значения. В наборе нет пропусков (NaN), так как он специально подготовлен для обучения.

### 4. Информация о типах данных

```python
print(df.info())
```
**Вывод:**
*   **Типы данных:** Все признаки имеют числовой тип данных (`float64`).
*   **Категориальные признаки:** Нет, все признаки количественные.
*   **Пропущенные значения:** Нет (`null count: 0`).

### 5. Исследование данных: Матрица корреляции

Построим тепловую карту корреляций, чтобы найти самые важные признаки.

```python
import seaborn as sns

corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Матрица корреляции')
plt.show()
```

Матрица покажет силу линейной связи между всеми признаками. Самый высокий коэффициент корреляции с целевой переменной будет выделен цветом.

### 6. Диаграмма рассеяния для самого важного признака

Найдем признак с максимальным коэффициентом корреляции (по модулю) и построим график.

```python
# Находим признак с максимальной корреляцией с target (игнорируя саму target)
corr_with_target = corr['target'].drop('target')
most_important_feature = corr_with_target.abs().idxmax()
corr_value = corr_with_target[most_important_feature]

print(f"Самый важный признак: {most_important_feature} (Корреляция: {corr_value:.3f})")

# Строим диаграмму рассеяния
plt.figure(figsize=(8, 6))
plt.scatter(df[most_important_feature], df['target'], alpha=0.6)
plt.title(f'Зависимость Target от {most_important_feature}')
plt.xlabel(most_important_feature)
plt.ylabel('Target (Прогрессирование болезни)')
plt.grid(True)
plt.show()
```

График покажет наличие линейной зависимости: при росте значения `s5` целевая переменная также имеет тенденцию к росту.

### 7. Формирование наборов X и y

Выберем 5 признаков с самой высокой корреляцией для обучения модели.

```python
# Выбираем топ-5 признаков по абсолютному значению корреляции
top_features = corr_with_target.abs().sort_values(ascending=False).head(5).index.tolist()
print("Топ-5 признаков:", top_features)

X = df[top_features] # Набор признаков (фичи)
y = df['target']     # Целевая переменная
```
*(Примечание: В зависимости от версии библиотеки, топ-5 могут немного меняться, но обычно это s5, bmi, s4, bp, s6).*

### 8. Разделение данных на обучающую и тестовую выборки

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(f"Размер обучающей выборки: {len(X_train)}, размер тестовой: {len(X_test)}")
```

Данные будут разделены в пропорции 75% на обучение и 25% на тестирование.

### 9. Обучение модели

Создадим и обучим модель линейной регрессии.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

### 10. Проверка модели и расчет метрик

Сделаем предсказания на тестовых данных и оценим качество модели.

```python
y_pred = model.predict(X_test)

# Расчет RMSE (среднеквадратичная ошибка)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
# Расчет R² (коэффициент детерминации)
r2 = r2_score(y_test, y_pred)

print(f"RMSE (Среднеквадратичная ошибка): {rmse:.2f}")
print(f"R² (Коэффициент детерминации): {r2:.2f}")
```

*   **RMSE:** Ошибка модели составляет примерно **50-60 единиц**. Это означает, что в среднем прогноз модели отклоняется от реального значения на эту величину. Учитывая масштаб целевой переменной (среднее значение около 150), это неплохой результат для простой линейной модели.
*   **R²:** Значение будет около **0.4 - 0.5**. Это означает, что модель объясняет около 40-50% разброса данных. Остальные 50% зависят от других факторов или шума. Модель имеет смысл, но не является идеальной.


##Ответы на вопросы:

1. Какие типы машинного обучения вы знаете?

Существует три основных типа машинного обучения:

- Обучение с учителем (Supervised Learning): Модель обучается на размеченном наборе данных, где для каждого примера известен правильный ответ (целевая переменная). Задача — научиться предсказывать этот ответ для новых данных. Наприер: Линейная регрессия (прогноз цены), классификация спама.
- Обучение без учителя (Unsupervised Learning): Модель работает с неразмеченными данными. Задача — найти скрытые структуры, закономерности или аномалии в данных. Напрммер: Кластеризация клиентов по поведению, понижение размерности данных.
- Обучение с подкреплением (Reinforcement Learning): Модель (агент) обучается, взаимодействуя со средой. Она совершает действия и получает «награды» или «штрафы» за них, стремясь максимизировать итоговую награду. Например: Обучение автопилота, игровые боты.

2. Чем отличается обучение с учителем и без учителя?

Главное отличие — в наличии «учителя», то есть целевой переменной (меток).

Обучение с учителем
- Данные размеченные (есть ответы)
- Цель предсказать значение или класс

Обучение без учителя
- Данные неразмеченные (нет ответов)
- Цель найти скрытые закономерности

3. Чем пакетное обучение отличается от динамического?

Это различие относится к тому, как и когда модель получает новые данные для обучения.

- Пакетное (Batch) обучение: Модель обучается на всем доступном наборе данных целиком (или большими порциями). Процесс обучения происходит редко, требует много ресурсов и останавливается. Чтобы обновить модель, её нужно переобучить заново на новом наборе данных.
- Динамическое (Online/Incremental) обучение: Модель обучается на данных по мере их поступления, по одному примеру или маленькими группами. Она постоянно обновляется и адаптируется к изменениям в реальном времени.

4. Чем обучение на основе образцов отличается от обучения на основе модели?

Различие в способе хранения информации о данных.

- Обучение на основе образцов (Instance-based): Модель не строит обобщенную функцию, а просто «запоминает» все обучающие примеры. Чтобы сделать предсказание для нового объекта, она ищет в памяти наиболее похожие (ближайшие) примеры из обучающей выборки и выдает усредненный ответ.Например: Метод k-ближайших соседей (k-NN).
- Обучение на основе модели (Model-based): В процессе обучения алгоритм строит явную математическую модель (функцию), которая описывает зависимость между признаками и целевой переменной. После обучения сами исходные данные можно удалить, а для предсказания используется только эта компактная модель. Например: Линейная регрессия, нейронные сети.

5. Что такое линейная регрессия?

Линейная регрессия — это метод обучения с учителем, используемый для решения задач регрессии (прогнозирования непрерывного числового значения).

Она предполагает, что зависимость между входными признаками (x) и выходной переменной (y) является линейной. Модель пытается найти такую прямую (в многомерном пространстве — гиперплоскость), которая лучше всего описывает данные, то есть минимизирует расстояние от точек данных до этой прямой.

Уравнение простой линейной регрессии:
y=ax+b.

6. Что такое градиентный спуск?

Градиентный спуск — это основной алгоритм оптимизации, используемый для поиска минимума функции ошибки (например, в линейной регрессии).

Представьте, что вы находитесь в горах в тумане и хотите спуститься в самую низкую долину. Вы не видите всей карты, но можете ощупью определить, в каком направлении склон идет вниз наиболее круто. Вы делаете шаг в этом направлении. Повторяя это действие много раз, вы в итоге окажетесь на дне долины.

В машинном обучении «горы» — это функция потерь, а «дно долины» — это набор параметров модели, при которых ошибка минимальна.

7. Как правильно обрабатывать категориальные признаки?

Машинное обучение работает с числами, поэтому категориальные признаки (текст: «мужчина», «Москва») нужно преобразовать. Основные методы:

- Label Encoding (Кодирование меток): Каждому уникальному значению присваивается целое число (например, "Москва" -> 0, "СПб" -> 1). Подходит только для порядковых данных.
- One-Hot Encoding (Унитарное кодирование): Для каждого уникального значения создается новый бинарный признак-столбец. Если у пассажира город "Москва", то в столбце City_Moscow будет 1, а в остальных 0. Стандартный подход для номинальных данных.
- Target Encoding (Кодирование по цели): Категории заменяются средним значением целевой переменной для этой категории. Требует осторожности для предотвращения утечки данных.

8. Что такое матрица корреляции?

Матрица корреляции — это таблица, которая показывает коэффициенты корреляции Пирсона между всеми парами числовых признаков в наборе данных.

- Диагональ матрицы всегда равна 1 (признак идеально коррелирует сам с собой).
- Значение 1 означает идеальную положительную линейную связь.
- Значение -1 означает идеальную отрицательную линейную связь.
- Значение 0 означает отсутствие линейной связи.Она используется для поиска взаимосвязей между признаками и отбора наиболее значимых из них для модели.

9. Что показывает метрика RMSE?

RMSE (Root Mean Squared Error) — среднеквадратичная ошибка. Она показывает среднюю ошибку предсказания модели в тех же единицах измерения, что и целевая переменная.

Чем меньше значение RMSE, тем точнее модель. Она сильно штрафует за большие ошибки из-за возведения в квадрат.

10. Что показывает метрика R²?

R² (Коэффициент детерминации) — это метрика, которая показывает, какую долю разброса данных объясняет модель. Она варьируется от −∞ до 1.

- R^2 = 1: Модель идеально описывает данные.
- R^2 = 0: Модель предсказывает просто среднее значение целевой переменной (не лучше константы).
- R^2 < 0: Модель работает хуже, чем простое предсказание среднего значения.

Например, R^2=0.75 означает, что 75% изменчивости целевой переменной объясняется моделью, а 25% остаются необъясненными (шум или влияние неучтенных факторов).
