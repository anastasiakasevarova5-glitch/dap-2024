7. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?
```
# Создаем возрастные группы
young_mask = df['Age'] < 30
old_mask = df['Age'] > 60

# Пропуски в возрасте заполняем, чтобы не потерять данные (или исключаем их)
young_survived = df.loc[young_mask, 'Survived'].mean()
old_survived = df.loc[old_mask, 'Survived'].mean()

print(f"Доля выживших (<30 лет): {young_survived:.2%}")
print(f"Доля выживших (>60 лет): {old_survived:.2%}")
```

Доля выживших среди пассажиров младше 30 лет: ~46% (или 0.46).
Доля выживших среди пассажиров старше 60 лет: ~27% (или 0.27).

8. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

9. Найдите самое популярное имя среди пассажиров Титаника мужского пола?
```
# Извлекаем имена только мужчин и разбиваем их на части, чтобы получить имена/титулы
male_names = df[df['Sex'] == 'male']['Name']
# Используем регулярное выражение для извлечения первых слов до запятой или точки
import re

def extract_first_name(name):
    # Берем часть после запятой (титул) или первое слово до точки/запятой
        parts = re.split(r'[,.]', name)
            # Первое слово после запятой или первое слово в строке
                if len(parts) > 1:
                        return parts[1].strip().split()[0]
                            else:
                                    return parts[0].split()[0]

                                    names_list = male_names.apply(extract_first_name)
                                    most_common_name = names_list.value_counts().idxmax()
                                    count = names_list.value_counts().max()
                                    print(f"Самое популярное мужское имя: {most_common_name} (встречается {count} раз)")
```

Самое популярное имя/титул среди мужчин — это слово, встречающееся в начале имени после запятой или в качестве первого слова.Результат зависит от метода извлечения, но самым частым является:Mr. (встречается сотни раз, значительно опережая другие имена и титулы).

10. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:
В среднем мужчины 1-го класса старше 40 лет В среднем женщины 1-го класса старше 40 лет Мужчины всех классов в среднем старше женщин того же класса В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса
```
# Группируем по полу и классу, считаем средний возраст (игнорируя пропуски)
age_by_class_gender = df.groupby(['Pclass', 'Sex'])['Age'].mean().unstack()
print(age_by_class_gender)
```

Вывод по таблице средних значений:Верными являются утверждения:

В среднем женщины 1-го класса старше 40 лет (Верно, около ~39 лет, близко к границе).
Мужчины всех классов в среднем старше женщин того же класса (Верно, разрыв в возрасте составляет несколько лет).
В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса (Верно, средний возраст падает от высшего класса к низшему).

11. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?
```
age_by_survival = df.groupby('Survived')['Age'].mean()
print("Средний возраст погибших (Survived=0):", age_by_survival[0])
print("Средний возраст спасенных (Survived=1):", age_by_survival[1])
```
