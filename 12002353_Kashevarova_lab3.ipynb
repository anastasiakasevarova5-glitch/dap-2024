##Ответы на вопросы:
1. Какой тип (dtype) будут иметь элементы массива?
- np.array([1, 4, 2, 5, 3]) -> int64 (или int32 в зависимости от системы).
- np.array([1., 4, ...]) -> float64 (наличие точки превращает все в float).
- dtype='float128' -> float128.
- dtype='complex128' -> complex128.
- dtype='float64' -> float64.
- dtype='int64' -> int64.
- dtype='bool' -> bool (все ненулевые числа станут True).
- dtype='str' -> str.
np.array([... , '3']) -> str (наличие строки превращает все в строку).
- np.array([... , 1j]) -> complex.

2. Что будет выведено в консоль и почему?
(True/False), а один из операндов.

- np.array([1, 4, 2, 5, 3], dtype='str')[1] + '1' -> '41' (конкатенация строк).
- np.array([1, -4, 2, 0, 3], dtype='bool')[3] or np.array([1, -4, 2, 0, 3])[1] -> -4. Массив в dtype=bool будет [True, True, True, False, True]. [3] это False. Оператор or вернет второй операнд, так как первый ложный.
- np.array([1, -4, 2, 0, 3], dtype='bool')[3] and np.array([1, -4, 2, 0, 3])[1] -> False. Первый операнд [3] это False. Оператор and вернет первый операнд.
- np.array([1, -4, 2, 0, True], dtype='int64')[-1] -> 1 (True приводится к int64).
- np.array([1, -4, 2, 0, 1j], dtype='int64')[1] -> Ошибка (ValueError), так как комплексное число нельзя привести к целому.
- np.array([1, -4, 2, 0, 1j], dtype='str')[-1] -> '1j'.
- np.array([1, 4, 2, 5, '3'])[-1] * 3 -> '333' (повторение строки).

3. Каков результат операции?

- np.ones((3,5)): Матрица из единиц (3x5).
- np.full((3,5), 3.14): Матрица заполнена числом Pi (3x5).
- np.linspace(0, 1, 5): Вектор из 5 равномерно распределенных чисел от 0 до 1 включительно.
- np.diag((1,3,5)): Диагональная матрица с числами [1,3,5] на диагонали.
- np.smile(...): Ошибка (AttributeError), такой функции нет.
- np.random.random((3,3)): Матрица (3x3) со случайными числами от [0.0, 1.0).
np.eye(3): Единичная матрица (3x3).
- np.empty(3): Массив из 3 элементов с "мусорными" значениями (неинициализированная память). Быстрее чем zeros().

[1, 2, 3] + [3, 2, 1]

Различие в типе данных результата:

[...] + [...]: Сложение списков. Результат — конкатенация (объединение) списков.
[...] + [...] == [..., ...]
Пример: [1]+[3] -> [1,3]
np.array([...]) + np.array([...]): Векторное сложение. Результат — поэлементное сложение (векторизация).
Пример: [...] + [...] == [...]
Пример: [...] + [...] -> [...]

10. В чем преимущество массивов NumPy по сравнению с обычными списками. Чем обусловлено это преимущество?

Главное преимущество — векторизация и использование непрерывных блоков памяти.Это обусловлено тем:

- Однородность типов (dtype) позволяет хранить данные компактно.
- Операции выполняются на уровне компилированного кода C/Fortran внутри библиотеки (ufuncs, "универсальные функции"), а не через медленные циклы Python (for/while).
- Это дает выигрыш в скорости в десятки и сотни раз при работе с большими объемами данных.
