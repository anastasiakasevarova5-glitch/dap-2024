##2. Ответы на вопросы
1. Типы ячеек в Jupyter Notebook
В Jupyter Notebook существует два основных типа ячеек:

Markdown — для написания и форматирования текста, вставки формул, списков, изображений и ссылок.
Code — для написания и выполнения программного кода (например, на Python).

2. Как с помощью Markdown сформировать иерархический список?

Для создания иерархического списка используются отступы (обычно 2 или 4 пробела) и разные маркеры.

Пример:
1. Элемент 1
2. Элемент 2
    - Элемент 2.1
    - Элемент 2.2
    1. Элемент 2.2.1
    2. Элемент 2.2.2
    3. Элемент 3

3. Как с помощью Markdown оформить цитату?

Цитата оформляется с помощью символа > (больше) в начале строки.
  Пример:
>Если называть Python заменой BASIC, то тогда и трансформер Optimus Prime — это только замена грузовика. — Cory Dodt

5. Сколько уровней заголовка существует в Markdown?

В Markdown существует 6 уровней заголовков:
1. # Заголовок 1 уровня
2. ## Заголовок 2 уровня
3. ### Заголовок 3 уровня
4. #### Заголовок 4 уровня
5. ##### Заголовок 5 уровня
6. ###### Заголовок 6 уровня

6. Шаги процесса анализа данных (подробно про один шаг)

Подготовка данных (Data Preparation) — это этап, на котором сырые данные преобразуются в формат, пригодный для анализа.Этот процесс включает:
- Очистку данных: удаление дубликатов, исправление ошибок, обработка пропущенных значений.
- Преобразование: изменение типов данных, нормализация или стандартизация числовых признаков.
- Инжиниринг признаков: создание новых колонок на основе существующих для улучшения качества модели.Качество этого этапа критически важно, так как ошибки на стадии подготовки приведут к неверным выводам на всех последующих шагах.
7. Типы данных

Полуструктурированные данные — это данные, которые не имеют строгой схемы (как таблицы в базах данных), но содержат внутренние теги или маркеры для разделения элементов.
- Примеры: JSON, XML, YAML.   - Особенность: В отличие от неструктурированных данных (текст, видео), их можно частично анализировать автоматически, но структура может меняться от записи к записи. Это делает их гибкими для обмена информацией между разными системами.
                    8. Что такое Jupyter Notebook, для чего используется?

Jupyter Notebook — это интерактивная веб-среда, которая позволяет объединять исполняемый код, текстовые пояснения (Markdown), математические формулы и визуализации в одном документе. Используется для анализа данных, машинного обучения, прототипирования и обучения.

9. Почему Python популярен для анализа данных?

Python стал стандартом де-факто благодаря сочетанию следующих факторов:
- Простота: Легко читаемый синтаксис позволяет быстро писать код.     - Экосистема: Огромное количество библиотек (Pandas, NumPy, Matplotlib, scikit-learn, TensorFlow).
- Сообщество: Большое количество готовых решений и документации.
- Интеграция: Легко взаимодействует с другими языками и базами данных.

7. Обязательно ли использовать Jupyter Notebook для анализа данных?

Необязательно, aнализ данных можно выполнять в любой среде разработки (IDE), например, PyCharm или VS Code, а также писать обычные Python-скрипты. Однако Jupyter Notebook удобен для интерактивного исследования данных и создания наглядных отчётов.
