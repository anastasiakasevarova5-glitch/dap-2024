## Задача 1: Линейные графики

Построим два графика функции $y = 2x + 3$ на одной системе координат.

```python
import matplotlib.pyplot as plt
import numpy as np

# Диапазоны x
x1 = np.linspace(-10, 0, 100)
x2 = np.linspace(0, 7, 100)

# Функция y = 2x + 3
y1 = 2 * x1 + 3
y2 = 2 * x2 + 3

plt.figure(figsize=(8, 6))

# Первый график: красный, штрихпунктирный
plt.plot(x1, y1, color='red', linestyle='-.', label='x ∈ [-10; 0]')

# Второй график: цвет (1.0, 0.2, 0.3), сплошной
plt.plot(x2, y2, color=(1.0, 0.2, 0.3), linestyle='-', label='x ∈ [0; 7]')

# Настройка осей и легенды
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Графики функции y = 2x + 3')
plt.legend()
plt.grid(True)
plt.show()
```
Графики построены в заданных диапазонах. Заметно, что при $x=0$ оба графика сходятся в одной точке $(0, 3)$. Штрихпунктирная линия (для отрицательных $x$) и сплошная линия (для положительных $x$) четко разграничивают области определения функции.


## Задача 2: Диаграмма рассеяния (Титаник)

Визуализируем зависимость стоимости билета от возраста пассажиров.

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('./data/titanic.csv')

# Удаляем строки с пропусками в Age и Fare
df_clean = df.dropna(subset=['Age', 'Fare'])

plt.figure(figsize=(10, 6))

# Выжившие: зеленые круги
survived = df_clean[df_clean['Survived'] == 1]
plt.scatter(survived['Age'], survived['Fare'],
            color='green', marker='o', alpha=0.6, label='Выжил')

# Погибшие: красные кресты
not_survived = df_clean[df_clean['Survived'] == 0]
plt.scatter(not_survived['Age'], not_survived['Fare'],
    color='red', marker='x', alpha=0.6, label='Погиб')

plt.xlabel('Возраст (Age)')
plt.ylabel('Стоимость билета (Fare)')
plt.title('Зависимость стоимости билета от возраста и выживаемости')
plt.legend()
plt.grid(True)
plt.show()
```

На диаграмме видно "облако" точек. Прямой зависимости между стоимостью билета и выживанием нет (точки перемешаны), однако можно заметить концентрацию "зеленых" точек (выживших) в области молодых женщин и детей, что подтверждает правило "сначала женщины и дети".


## Задача 3: Матрица диаграмм рассеяния (Iris)

Построим матрицу всех попарных зависимостей для признаков ириса.

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

# Цвета для видов
colors = ['blue', 'green', 'red']
targets = iris.target_names

pd.plotting.scatter_matrix (df_iris.iloc[:, :4],
                            c=df_iris['target'],
                            figsize=(12, 10),
                            marker='o',
                            hist_kwds={'bins': 20},
                            s=60,
                            alpha=0.8,
                            cmap=plt.cm.get_cmap('viridis', 3))

# Добавляем легенду вручную (сложнее в scatter_matrix)
handles = [plt.Line2D([0], [0], marker='o', color='w', label=targets[i],
                      markerfacecolor=colors[i], markersize=10) for i in range(3)]
plt.gcf().axes[-1].legend(handles=handles, title="Виды")
plt.suptitle("Матрица диаграмм рассеяния для набора данных Iris", y=1.02)
plt.show()
```

Матрица наглядно показывает разделение видов ирисов. Видно, что `setosa` (синий) хорошо отделяется от `versicolor` (зеленый) и `virginica` (красный) по признакам длины чашелистика. Виды `versicolor` и `virginica` частично пересекаются.


## Задача 4: Гистограммы (Титаник)

Сравним распределение возраста всех пассажиров и распределение стоимости билетов только для женщин.

```python
import pandas as pd

df = pd.read_csv('./data/titanic.csv')
df_clean = df.dropna(subset=['Age', 'Fare'])

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Гистограмма возраста (все пассажиры)
ax[0].hist(df_clean['Age'], bins=np.arange(0, 85, 5), color='skyblue', edgecolor='black')
ax[0].set_xlabel('Возраст')
ax[0].set_ylabel('Количество пассажиров')
ax[0].set_title('Распределение пассажиров по возрасту')
ax[0].grid(True)

# Гистограмма стоимости билетов (только женщины)
female_fares = df_clean[df_clean['Sex'] == 'female']['Fare']
ax[1].hist(female_fares, bins=25, color='pink', edgecolor='black')
ax[1].set_xlabel('Стоимость билета')
ax[1].set_ylabel('Количество пассажиров')
ax[1].set_title('Распределение женщин по стоимости билетов')
ax[1].grid(True)

fig.suptitle('Гистограммы распределения данных Титаника', fontsize=16)
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Распределение возрастов бимодальное: есть пики для детей/подростков (~20 лет) и взрослых (~30 лет). Распределение стоимости билетов у женщин сильно скошено вправо: большинство женщин платили мало за билет (левый пик), но были и те, кто заплатил очень много (длинный "хвост").


## Задача 5: Столбчатая диаграмма (Средняя стоимость билета по классам)

```python
import pandas as pd

df = pd.read_csv('./data/titanic.csv')
mean_fare_by_class = df.groupby('Pclass')['Fare'].mean()

mean_fare_by_class.plot(kind='bar', color=['navy', 'darkorange', 'darkgreen'])
plt.xlabel('Класс каюты (Pclass)')
plt.ylabel('Средняя стоимость билета')
plt.title('Средняя стоимость билета в зависимости от класса')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Средняя стоимость билета строго убывает с повышением класса каюты: **1-й класс > 2-й класс > 3-й класс**. Это логично отражает социально-экономический статус пассажиров.


## Задача 6: Столбчатая диаграмма с накоплением (Класс каюты и Пол)

```python
import pandas as pd

df = pd.read_csv('./data/titanic.csv')
# Считаем количество мужчин и женщин в каждом классе
stacked_counts = df.groupby(['Pclass', 'Sex']).size().unstack(fill_value=0)

stacked_counts.plot(kind='bar', stacked=True, figsize=(8,6), colormap='Pastel2')
plt.xlabel('Класс каюты (Pclass)')
plt.ylabel('Количество пассажиров')
plt.title('Распределение пассажиров по классу и полу')
plt.legend(title='Пол')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Во всех трех классах мужчин было больше, чем женщин. Однако разрыв сокращается от 3-го класса к 1-му: в первом классе соотношение полов наиболее сбалансировано.


## Задача 7: Круговая диаграмма (Виды Ирисов)

```python
from sklearn.datasets import load_iris

iris = load_iris()
target_counts = pd.Series(iris.target).value_counts().sort_index()
labels = iris.target_names

explode = [0.1 if i == target_counts.idxmax() else 0 for i in range(len(labels))]
# Если два вида имеют одинаковую макс. долю, explode будет у первого из них.

fig, ax = plt.subplots(figsize=(6,6))
wedges, texts = ax.pie(target_counts.values,
                       labels=[f"{l}\n({c})" for l,c in zip(labels, target_counts.values)],
                       autopct='%1.1f%%',
                       startangle=90,
                       explode=explode,
                       shadow=True,
                       wedgeprops=dict(width=0.4))
ax.legend(wedges, labels,
          title="Виды",
          loc="center left",
          bbox_to_anchor=(1.25, 0.5))
ax.set_title("Доли видов ирисов в наборе данных")
plt.show()
```

В наборе данных Iris виды распределены равномерно: по **50** экземпляров каждого вида (`setosa`, `versicolor`, `virginica`). Сектор самого крупного вида визуально выделен из центра диаграммы.


## Задача 8: Гистограмма возраста с помощью Seaborn

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('./data/titanic.csv')
df_clean = df.dropna(subset=['Age'])

sns.set_style("whitegrid")
fig_dims = (8, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.histplot(data=df_clean, x='Age', bins=25, kde=True, color='teal', ax=ax)
ax.set_title('Распределение возраста пассажиров Титаника')
ax.set_xlabel('Возраст')
ax.set_ylabel('Количество пассажиров')
sns.despine() # Убираем верхнюю и правую границы графика для красоты Seaborn
```

Гистограмма подтверждает выводы из задачи 4: распределение возрастов имеет два пика. Линия KDE (*Kernel Density Estimate*) сглаживает данные и делает бимодальность еще более очевидной. Видно преобладание молодых пассажиров.


##Ответы на вопросы:

1. Для чего используется диаграмма рассеяния?

Диаграмма рассеяния (scatter plot) используется для визуализации взаимосвязи между двумя числовыми переменными.

Она помогает:

- Оценить, существует ли корреляция (зависимость) между признаками (например, чем больше возраст, тем выше стоимость билета).
- Выявить выбросы — точки, которые сильно отклоняются от общей массы данных.
- Заметить нелинейные зависимости или кластеризацию данных.


2. Что такое гистограмма, чем он отличается от столбчатой диаграммы?

- Гистограмма показывает распределение одной числовой переменной. Ось X делится на интервалы (бины), а высота столбца показывает, сколько значений попало в этот интервал. Столбцы гистограммы всегда стоят вплотную друг к другу.
- Столбчатая диаграмма (bar chart) сравнивает значения разных категорий. Ось X содержит названия категорий (например, «Класс 1», «Класс 2»), а высота столбца — значение для этой категории. Между столбцами всегда есть зазор.

3. Как задать цвет линии в линейном графике Matplotlib?

Цвет линии задается с помощью параметра color (или его сокращения c) в функции plt.plot().

Примеры:




```
plt.plot(x, y, color='red')        # По названию цвета
plt.plot(x, y, color='#FF0000')    # По шестнадцатеричному коду
plt.plot(x, y, color=(1.0, 0, 0))  # В формате RGB (от 0 до 1)
```


4. Чем класс plt.Figure отличается от объекта plt.Axes?
Matplotlib не позволяет управлять типом маркера scatter графика. Да или нет?

Это фундаментальное различие в структуре Matplotlib:

- Figure (Фигура) — это весь холст (окно или картинка), на котором размещаются графики. Он является контейнером для всех элементов.
- Axes (Оси) — это конкретный график внутри фигуры. На одной фигуре (Figure) может быть несколько осей (Axes) — например, при создании подграфиков.

Проще говоря: Figure — это страница альбома, а Axes — это отдельный рисунок на этой странице.

6. Matplotlib единственная библиотека для построения графиков в Python. Да или нет?

Нет. Существует множество других библиотек:

- Seaborn (построена на базе Matplotlib, для красивых статистических графиков).
- Plotly (для интерактивных веб-графиков).
- Bokeh (для интерактивной визуализации больших данных).
- Altair (декларативная библиотека на основе Vega-Lite).

7. Matplotlib не позволяет сохранить график в виде файла, нужно делать скриншот?

Это утверждение неверно. Библиотека Matplotlib позволяет детально настраивать маркеры в диаграммах рассеяния с помощью параметров marker, s (размер), alpha (прозрачность) и других.

10. Как на одном холсте может располагаться не более 16 диаграмм, это обусловлено тем что размер диаграммы нельзя изменить и более мелкие графики попросту не будут видны.

Это утверждение является верным объяснением того, почему на больших матрицах подграфиков (например, 4x4 = 16) не рекомендуется делать их еще больше.

Если разместить на одном холсте слишком много маленьких графиков (например, 20x20 = 400), то:

- Размер каждого отдельного графика станет микроскопическим.
- Различить тренды, выбросы или даже прочитать подписи будет невозможно.
- График превратится в нечитаемое «пятно».

Поэтому ограничение по количеству и размеру — это вопрос читаемости и здравого смысла.
