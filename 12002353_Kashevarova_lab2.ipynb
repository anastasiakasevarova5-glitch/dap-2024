```python
class Solution1(object):
  def twoSum(self, nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: List[int]
    """
    # Используем словарь для хранения числа и его индекса.
    # Это позволяет найти пару за один проход (O(n)).
    seen = {}
    for i, num in enumerate(nums):
      complement = target - num
      if complement in seen:
        return [seen[complement], i]
      seen[num] = i
    return []
# Проверка
sol1 = Solution1()
print(sol1.twoSum([2, 7, 11, 15], 9)) # [0, 1]
```

```python
class Solution2(object):
  def isPalindrome(self, x):
    """
    :type x: int
    :rtype: bool
    """
    if x < 0:
      return False

    s = str(x)
    return s == s[::-1]

# Проверка
sol2 = Solution2()
print(sol2.isPalindrome(121))   # True
print(sol2.isPalindrome(-121))  # False
```

```python
class Solution3(object):
  def romanToInt(self, s):
    """
    :type s: str
    :rtype: int
    """

    roman = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}

    total = 0
    prev_value = 0

    # Проходим по строке справа налево.
    for char in s[::-1]:
      value = roman[char]
      if value < prev_value:
        total -= value
      else:
        total += value
      prev_value = value

    return total

# Проверка
sol3 = Solution3()
print(sol3.romanToInt("III"))      # 3
print(sol3.romanToInt("LVIII"))    # 58
print(sol3.romanToInt("MCMXCIV"))  # 1994
```

```python
class Solution4(object):
  def longestCommonPrefix(self, strs):
    """
    :type strs: List[str]
    :rtype: str
    """

    if not strs:
      return ""

    # Берем первую строку как эталон.
    prefix = strs[0]

    for i in range(len(prefix)):
      char = prefix[i]
      for string in strs[1:]:
        if i >= len(string) or string[i] != char:
          return prefix[:i]
    return prefix

# Проверка
sol4 = Solution4()
print(sol4.longestCommonPrefix(["flower","flow","flight"])) # "fl"
```

```python
class ListNode(object):
  def __init__(self, val=0, next=None):
    self.val = val
    self.next = next

class Solution5(object):
  def mergeTwoLists(self, list1, list2):
    """
    :type list1: Optional[ListNode]
    :type list2: Optional[ListNode]
    :rtype: Optional[ListNode]
    """
    dummy = ListNode()
    current = dummy

    while list1 and list2:
      if list1.val < list2.val:
        current.next = list1
        list1 = list1.next
      else:
        current.next = list2
        list2 = list2.next
      current = current.next

    current.next = list1 if list1 else list2

    return dummy.next

# Проверка (создание списков и вывод результата)
# Список 1: 1 -> 3 -> 5
l1 = ListNode(1, ListNode(3, ListNode(5)))
# Список 2: 2 -> 4 -> 6
l2 = ListNode(2, ListNode(4, ListNode(6)))
sol5 = Solution5()
merged = sol5.mergeTwoLists(l1, l2)
# Вывод результата в виде списка Python для наглядности:
result = []
while merged:
  result.append(merged.val)
  merged = merged.next
print(result) # [1, 2, 3, 4, 5, 6]
```

```python
class Solution(object):
  def lengthOfLongestSubstring(self, s):
    """
    :type s: str
    :rtype: int
    """
    char_index_map = {}
    max_length = 0
    # Начало текущего окна без повторов.
    start_window = 0

    for end_window, char in enumerate(s):
      if char in char_index_map and char_index_map[char] >= start_window:
        start_window = char_index_map[char] + 1
      # Обновляем индекс текущего символа.
      char_index_map[char] = end_window
      # Обновляем максимальную длину.
      max_length = max(max_length, end_window - start_window + 1)

    return max_length

# Проверка
sol6 = Solution()
print(sol6.lengthOfLongestSubstring("abcabcbb")) # 3 ("abc")
print(sol6.lengthOfLongestSubstring("bbbbb"))    # 1 ("b")
```

1. Что такое динамическая типизация и чем она отличается от статической?

Динамическая типизация (Python) означает, что тип переменной определяется во время выполнения программы (x = 5 делает x целым числом). В статической типизации (C++, Java) тип переменной должен быть объявлен заранее и не может измениться (int x;).

Плюс Python: Гибкость и скорость написания кода.
Минус Python: Ошибки типов могут возникнуть только при запуске программы.
Плюс C++: Ошибки типов ловятся компилятором до запуска.
Минус C++: Нужно явно указывать типы.(Примечание: В Python есть опциональные подсказки типов — type hints).

2. Чем отличается цикл for в Python от цикла for в С++?

В C++ цикл for — это классический цикл со счетчиком:for (int i=0; i<10; i++).В Python цикл for — это цикл по итерируемому объекту (список, строка). Он перебирает элементы коллекции напрямую:for item in my_list:.Если нужен счетчик в Python, используют функцию range():for i in range(10):.

5. Что такое лямбда функции и для чего они предназначены?

Лямбда-функция — это анонимная (без имени) функция, определенная в одну строку.Синтаксис: lambda аргументы: выражение.Они предназначены для создания простых функций «на лету», которые используются как аргументы для других функций (например, key в sorted() или filter()).Пример:

python
  numbers = [4, -7, 2]
  sorted_numbers = sorted(numbers, key=lambda x: abs(x))

Здесь lambda x: abs(x) говорит функции сортировки сортировать по модулю числа.

9. Как определить статический метод класса? Для чего может понадобиться статический метод?

Статический метод определяется декоратором @staticmethod.Он не получает первым аргументом ни экземпляр (self), ни класс (cls).Он ведет себя как обычная функция, но логически принадлежит классу.Используется для вспомогательных функций внутри модуля класса.

python
class MathUtils:
  @staticmethod # Нет self/cls в аргументах!
  def add_five(x):
    return x + 5

MathUtils.add_five(10) # Работает без создания объекта!
                                  10. Чем staticmethod отличается от classmethod?

- @staticmethod: Не получает никаких специальных аргументов. Не знает о классе или объекте. Просто функция внутри пространства имен класса.
- @classmethod: Получает первым аргументом сам класс (cls). Может создавать новые экземпляры класса или обращаться к атрибутам самого класса (не экземпляра).Пример использования classmethod как альтернативного конструктора:
python
class User:
  def __init__(self, name):
  self.name = name

  @classmethod # cls здесь будет User!
  def from_string(cls, data_str):
    name = data_str.split(":")[0]
    return cls(name) # Создаем новый User!

11. Как определить собственный класс исключений Python? Как выбросить и перехватить исключение?

Создается класс-наследник от встроенного Exception.Выбросить исключение можно командой raise.Перехватить — конструкцией try/except.

python
class MyCustomError(Exception): pass

def check_age(age):
  if age < 0:
    raise MyCustomError("Возраст не может быть отрицательным!")
  print(f"Вам {age} лет.")

try:
  check_age(-5)
except MyCustomError as e:
  print(f"Поймали ошибку: {e}")

12. Что такое кортеж (tuple), для чего он используется?

Кортеж — это неизменяемый (immutable) список. Определяется круглыми скобками (1, "a", True).Используется для хранения разнородных данных (записей), которые не должны меняться.

- Неизменяемость: Защита данных от случайного изменения.
- Хешируемость: Кортежи можно использовать как ключи в словарях (dict), а списки — нельзя.

- Возврат нескольких значений: Как упоминалось выше, функции возвращают кортежи.
